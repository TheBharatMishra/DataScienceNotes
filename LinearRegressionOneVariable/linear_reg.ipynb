# Linear Regression in one variable


Firstly we import all the modules and libraries needed. Here we used pandas for data cleaning and reading files. Numpy to work with multidimensional arrays and mathematical functions. Matplotlib for graphing the data. Sklearn provides us with machine learning algorithms and an easier way to work with them. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

Now read the csv file using pandas function and return some data using head() function (If we return the whole data it may take up a lot of apce and might not look desirable)

```python
df = pd.read_csv('per_capita_income.csv')
df.head()
```

Here we plot the graph taking year value as X and per capita income as Y.

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Per Capita Income(US$)')
plt.scatter(df.year,df['per capita income (US$)'],color='orange',marker='+')
```

Applying the linear regression algotrithm on the given data.

```python
reg= linear_model.LinearRegression()
reg.fit(df[['year']],df['per capita income (US$)'])
```

```python

p = pd.read_csv('per_capita_income.csv')
prediction_capita = reg.predict(p[['year']])

p['Predicted Per Capita Income(US$)'] = prediction_capita

p.to_csv('prediction.csv')

new_prediction= pd.read_csv('prediction.csv')

new_prediction.head()

%matplotlib inline
plt.scatter(new_prediction.year, new_prediction['per capita income (US$)'],marker='+',color='red')
plt.plot(new_prediction.year,new_prediction['Predicted Per Capita Income(US$)'],color='blue')

```
